```python
import os
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

Here I will be doing some webscraping using BeautifulSoup

```python
url = 'https://www.murphysmultiverse.com/marvel-studios-movie-runtimes/'

headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36'}

request = requests.get(url, headers = headers) # Need to include more information to access this website 
request
```

```python
content = request.text
content
```

```python
soup = BeautifulSoup(content)
```

```python
text = soup.find_all('p')[3:-5]
```

```python
info = [val.text for val in text]
info
```

```python
info_copy = info
bad_entries = []
for key, value in enumerate(info_copy): # Website did not have information organized in a table, so some of the entries are just sentences talking about each film phase. 
    if len(value.split('Release')[0]) > 50: # If the entry does not contain a release date, then it is not a film entry, and the split will do nothing, leaving a phrase longer than 50 characters
        bad_entries.append(key) # Keeping track of which entry indices are the bad ones 

for i in range(0, len(bad_entries)): # removing the bad entries based on index 
    del info_copy[bad_entries[i] - i]

len(info_copy)
```

The length still does not seem right based on the number of movies I should have. 

```python
info_copy # checking to see whats going on 
```

For some reason the entries for Thor: The Dark World and Iron Man 3 are on the same line. I went back to look at the data from bs4, and for some reason the entries came across on the same paragraph. It will need to be fixed.

```python
thor_darkworld = 'Thor' + info_copy[6].split('Thor')[1] # Manually splitting the string causing troubles 
iron_man3 = info_copy[6].split('Thor')[0]
index_to_replace = 6

updated_list = []
newOrder = False

for i in range(0, len(info_copy) + 1): # Replacing the strings and reassigning each value 
    if i == index_to_replace:
        updated_list.append(iron_man3)
    elif i == index_to_replace + 1:
        updated_list.append(thor_darkworld)
        newOrder = True
    elif newOrder == True:
        updated_list.append(info_copy[i - 1])
    else:
        updated_list.append(info_copy[i])
    
len(updated_list) 
```

The new length looks correct. 

```python
names = []
row_values = []
for entry in updated_list:
    names.append(entry.split('Release')[0].strip())
    row_values.append(entry.split('Runtime')[1].strip())

row_values # checking to see how it looks 
```

For some reason there is a missing colon on the 3rd to last entry. 

```python
row_values[-3] = ': 2 hours 5 minutesNumber of stingers: 2' # Fixed it!
row_values 
```

```python
end_stripped = [] 

for element in row_values: # Do not need number of stingers, so they will be dropped in order to get the run time 
    end_stripped.append(element.split('Number')[0].strip())
    
end_stripped 
```

```python
front_stripped = []

for ele in end_stripped: # Removing the colons 
    front_stripped.append(ele.split(': ')[1].strip())

front_stripped
```

```python
final_product = {'Film' : names, 'Runtime' : front_stripped}
```

```python
df = pd.DataFrame(final_product)
df # Missing the newest Deadpool movie
```

This website was missing the values for the newest marvel movie, Deadpool and Wolverine, so I will need to go to another website to get that information.

```python
new_url = 'https://www.imdb.com/title/tt6263850/'

new_request = requests.get(new_url, headers = headers)
new_request
```

```python
new_content = new_request.text
new_content
```

```python
soup = BeautifulSoup(new_content)
```

```python
new_text = soup.find_all('div', class_ = 'sc-1f50b7c-0 iPPbjm')
```

```python
converted_text = [txt.text for txt in new_text]
```

```python
converted_text
```

```python
deadpool_title = converted_text[0].split('2024')[0] # Isolating the length of the film 
deadpool_title
```

```python
deadpool_runtime = converted_text[0].split('2024R')[1]
deadpool_runtime
```

```python
deadpool_runtime = '2 hours 8 minutes' # need to convert h and m to hours and minutes respectively 
```

```python
final_product['Film'].append(deadpool_title)
final_product['Runtime'].append(deadpool_runtime)
final_product
```

```python
final_df = pd.DataFrame(final_product)
final_df
```

```python
times = final_product['Runtime'] # Just realized that I need to convert hours and minutes to just minutes for future analysis 
```

```python
hours = []
minutes = []

for time in times:
    hours.append(int(time.split('h')[0].strip())) # Storing hours and minutes 
    minutes.append(int(time.split('minute')[0].split(' ')[2].strip()))
```

```python
hours_to_min = [60 * hr for hr in hours] # Converting hours to minutes 
hours_to_min
```

```python
hours_plus_mins = list(map(np.add, hours_to_min, minutes))
hours_plus_mins # adding each of the entries together 
```

```python
final_product['Runtime'] = hours_plus_mins
final_product
```

```python
df = pd.DataFrame(final_product)
```
