```python
import os 
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = os.getcwd()
```

```python
film_ratings = pd.read_csv(path + '/Ratings.csv')
film_ratings
```

```python
film_runtimes = pd.read_csv(path + '/Runtime.csv')
film_runtimes
```

```python
film_earnings = pd.read_csv(path + '/Marvel_Movie_Economics.csv')
film_earnings
```

```python
earnings_runtime = film_earnings.merge(film_runtimes, on = 'Film')
earnings_runtime
```

```python
df = earnings_runtime.merge(film_ratings, on = 'Film')
df['Net_Revenue'] = df['Box office gross-Worldwide'] - df['Budget'] # New column containing net revenue from each film
df
```

```python
df.to_csv(path + '/Merged_DF.csv', index = False)
```

```python
df.set_index('Film', inplace = True)
df
```

```python
new_df = df.drop(columns = ['CinemaScore', 'U.S. release date', 'Box office gross-U.S. and Canada', 'Box office gross-Other territories']) # only using columns with numerical data 
new_df
```

```python
sns.barplot(x = new_df.index, y = new_df['Box office gross-Worldwide'])
plt.xticks(rotation = 90)
plt.title('Gross Revenue Per Film')
plt.ylabel('Amount (USD)')
```

```python
plt.scatter(new_df.index, new_df['Budget'], label = 'Budget', color = 'red')
sns.scatterplot(x = new_df.index, y = new_df['Box office gross-Worldwide'], label = 'Gross Revenue', color = 'green')
plt.xticks(rotation = 90)
plt.title('Comparing Gross Revenue and Budget')
plt.ylabel('Amount (USD)')
```

```python
sns.lineplot(x = new_df.index, y = new_df['Net_Revenue'])
plt.xticks(rotation = 90)
plt.title('Net Revenue Per Movie')
plt.ylabel('Net Revenue (USD)')
plt.xlabel('Film')
```

```python
corr_data = new_df.corr() 
sns.heatmap(corr_data)
```

```python
sns.scatterplot(x = new_df['Rotten Tomatoes'], y = new_df['Metacritic']) # Metacritic and Rotten Tomatoes appear to be the two columns most highly correlated in the dataframe
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(new_df['Rotten Tomatoes'], new_df['Metacritic'])
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train.values.reshape(-1, 1), y_train.values)
```

```python
model_prediction = lr.predict(x_test.values.reshape(-1,1))
```

```python
plt.plot(x_test, model_prediction, label = 'Linear Regression Model', color = 'red')
sns.scatterplot(x = x_test, y = y_test, label = 'Actual Values', color = 'blue')
plt.title('Comparing Linear Regression Model Against Actual Values')
```

```python
lr.score(x_test.values.reshape(-1,1), y_test.values)
```

Not the best R squared value, but I would say it is decent for this sample.
