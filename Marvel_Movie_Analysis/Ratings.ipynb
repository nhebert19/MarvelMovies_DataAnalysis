```python
import requests
import os
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_films'

request = requests.get(url)
request
```

```python
content = request.text
content
```

```python
soup = BeautifulSoup(content)
```

```python
soup.find_all('table')
```

```python
table = soup.find_all('table', class_ = 'wikitable plainrowheaders sortable')[1]
table
```

```python
header = table.find_all('th')[0:7]
column_names = [name.text.strip() for name in header]
column_names.pop()
column_names.pop(1)
column_names.pop(1)
column_names
```

```python
all_row_data = []
rows = table.find_all('tr')[3:]
for row in rows:
    title = row.find_all('th')
    title_info = [val.text.strip() for val in title]
    tmp1 = title_info
    row_data = row.find_all('td')
    info = [data.text.strip() for data in row_data]
    tmp2 = info
    tmp1 = tmp1 + tmp2
    all_row_data.append(tmp1)

all_row_data
```

```python
for list in all_row_data: # remove the last column (too many missing values)
    list.pop()
    
```

```python
all_row_data = [list for list in all_row_data if len(list) != 0] # Remove empty lists
all_row_data
```

```python
copy_list = []

for i, j in enumerate(all_row_data):
    tmp = []
    for key, value in enumerate(all_row_data[i]): # There is a bunch of extra information not needed, so I will be removing it using this loop 
        if '(' in value: # Remove the (# of reviews) 
            all_row_data[i][key] = all_row_data[i][key][0:3].strip()
        elif '[' in value: # Remove the [reference number]
            all_row_data[i][key] = all_row_data[i][key][:-5]
            
            
all_row_data   
```

```python
df = pd.DataFrame(all_row_data, columns = column_names)
df
```

```python
rotten_tomatoes = df['Rotten Tomatoes']
rotten_tomatoes
```

```python
def remove_percent(x): # Need to remove % for future analysis 
    return x[:-1]
```

```python
temp = rotten_tomatoes.apply(remove_percent)
temp
```

```python
df['Rotten Tomatoes'] = temp
df 
```
